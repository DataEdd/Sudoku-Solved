# Ground Truth Annotation Tool

**Workflow for each image:**
1. Run the annotation cell — the image will appear
2. **Click 4 corners** on the Sudoku grid in order: **Top-Left → Top-Right → Bottom-Right → Bottom-Left**
3. The warped (perspective-corrected) grid will appear
4. **Type the 9x9 grid** into the text box — use `0` for empty cells, one row per line
5. Click **Save** to store this annotation
6. Move to the next image by running the cell again

All annotations are saved to `evaluation/ground_truth_annotated.json`.
